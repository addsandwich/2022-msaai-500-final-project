```python
n = 100
a = len(data)
b = int(len(data)/n)
```

```python
res = []
for i in range(b):
    x=0
    temp = n if a >= i*n+n else a-i*n
    for j in range(temp):
        x+= data.iloc[n*i+j]['Price']
    res.append(int(x/temp))
res.sort()
```

```python
res2 = res[:-1]
plt.plot(res2)
print(b)
```

```python
import numpy
μ = numpy.mean(res2, axis=0)
σ = numpy.std(res2, axis=0)
print(μ)
print(σ)
```
