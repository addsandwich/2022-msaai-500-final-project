```python
# Importing important libraries
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Get Path
path = 'C:/Users/chris/Documents/School/Masters/zz_GIT/2022-msaai-500-final-project/data/sanitized/FilterData3.csv'

# Reading the dataset
df = pd.read_csv(path)
display(df)
```

```python
X = df[['Levy','MakeModel','Prod_year','Category','Leather_interior','Fuel_type','Mileage','Engine_volume','Cylinders','Gear_box_type','Drive_wheels','Doors','Wheel','Color','Airbags']]
```

```python
X = pd.get_dummies(data=X, drop_first=True)
X.head()
```

```python
Y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=101)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
# print the intercept
print(model.intercept_)
```

```python
coeff_parameter = pd.DataFrame(model.coef_,X.columns,columns=['Coefficient'])
coeff_parameter
```

```python
predictions = model.predict(X_test)
predictions
```

```python
sns.regplot(y_test,predictions)
```
